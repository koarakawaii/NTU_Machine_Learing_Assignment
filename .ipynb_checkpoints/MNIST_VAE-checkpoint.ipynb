# Autoencoder and Variational Autoencoder Implemented on MNIST Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from ipywidgets import FloatSlider, IntSlider, interact_manual

%env KERAS_BACKEND = tensorflow
%matplotlib inline
```

```python
from keras.datasets import mnist
from keras.models import Model
from keras.layers import Input, Lambda
from keras.layers import Dense
from keras.optimizers import Adam
from keras.losses import MSE
from keras import metrics
from keras.datasets import mnist
from keras.utils import np_utils
from keras import backend as K
```

## 對資料做預處理

```python
(x_train0,y_train0),(x_test0,y_test0) = mnist.load_data()
```

```python
print('Total %d data for training with shape (%d,%d) for each element.' %x_train0.shape)
print('Total %d data for testing with shape (%d,%d) for each element.' %x_test0.shape)
```

```python
#index = np.random.randint(len(x_train))
plt.imshow(x_train0[5673], cmap = 'Greys')
plt.show()
```

```python
x_train = x_train0.reshape(len(x_train0),-1)
x_test = x_test0.reshape(len(x_test0),-1)
y_train = np_utils.to_categorical(y_train0, 10)
y_test = np_utils.to_categorical(y_test0, 10)

x_train -= x_train.min()
x_train = x_train/x_train.max()
x_test -= x_test.min()
x_test = x_test/x_test.max()
```

```python
print('Total %d data for training with shape %d for each element.' %x_train.shape)
print('Total %d data for testing with shape %d for each element.' %x_test.shape)
```

## 先嘗試普通的 Autoencoder ，不過把 Dense 層的神經元變多，同時 activation function 改用 relu

```python
x = Input(shape=(784,))
enc_1 = Dense(256, activation='relu')
enc_2 = Dense(2, activation='relu')

h = enc_2(enc_1(x))
h_prime = Input(shape=(2,))

dec_2 = Dense(256, activation='relu')
dec_1 = Dense(784, activation='relu')

x_prime = dec_1(dec_2(h_prime))
```

```python
AE_total = Model(x, dec_1(dec_2(h)))
AE_total.summary()
```

### 訓練完權重後 load_weights

```python
AE_total.compile(loss='mse', optimizer=Adam(), metrics=['acc'])
AE_total.load_weights('./VAE_MNIST_model/Model_1st_weights.h5')
#Model_1st_history = AE_total.fit(x_train, x_train, batch_size=1024, epochs=128)
```

```python
#Model_1st_json = AE_total.to_json()
#open('./VAE_MNIST_model/Model_1st','w').write(Model_1st_json)
#AE_total.save_weights('./VAE_MNIST_model/Model_1st_weights.h5')
```

### 把 encoder 跟 decoder 分開

```python
AE_encoder = Model(x,h)
AE_encoder.summary()
```

```python
AE_decoder = Model(h_prime, x_prime)
AE_decoder.summary()
```

### 確定兩者可訓練的參數加起來跟兩者串聯起來的總參數一樣多

```python
201474+202256 == 403730
```

### 可視化 encoded 後的二維變數，可以看到 activation function 用 relu， 就算只用 Autoencoder 也做得不差， 0, 1, 2, 6, 7都分得蠻清楚的。變數的範圍都在第一象限，感覺是因為用 relu 造成的(非負的 activation)?

```python
sample = np.random.randint(len(x_test), size=len(x_test)//2)
encoder_vis = AE_encoder.predict(x_test[sample])
plt.scatter(encoder_vis[:,0], encoder_vis[:,1], c = y_test0[sample], cmap='tab10')
plt.colorbar()
plt.xlabel('Variable 1')
plt.ylabel('Variable 2')
plt.show()
```

### 可視化指定的二維平面上各點被 decoded 後的結果，可以看到結果跟 encoded 後的分佈(上圖)是類似的

```python
n = 32
pixel_side_each = 28
pixel_side_total = pixel_side_each*n
fig = np.zeros((pixel_side_total,pixel_side_total))
grid_x = np.linspace(0, 30, n)
grid_y = np.linspace(0, 35, n)

for i in range(0,n):
    for j in range(0,n):
        input_data = np.array([[grid_x[i], grid_y[j]]])
        temp = AE_decoder.predict(input_data)
        temp = 255*(temp-temp.min())/temp.max()
        #print(temp.max())
        temp = temp.reshape(28,28)
        fig[(n-j-1) * pixel_side_each: (n-j) * pixel_side_each, #from up to down
            i * pixel_side_each: (i + 1) * pixel_side_each] = temp # from left to right
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(fig, cmap='Greys')
plt.axis('off')
plt.show()
```

## 改用 Variational Autoencoder 看看
### 定義好抽樣的 layer 跟自訂的 loss function

```python
def sampling(args):
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape=(2,), mean=0., stddev=1)
    return z_mean + K.exp(z_log_var / 2) * epsilon
def vae_loss(x, x_recon):  
    recovery_loss = 784 * metrics.binary_crossentropy(x, x_recon)
    kl_loss = - 0.5 * K.sum(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis=-1)
    return recovery_loss + kl_loss
```

```python
enc_1_VAE = Dense(256, activation='relu')
enc_2_VAE_mean = Dense(2)
enc_2_VAE_log_var = Dense(2)
z_mean = enc_2_VAE_mean(enc_1_VAE(x))
z_log_var = enc_2_VAE_log_var(enc_1_VAE(x))

NS_sampling = Lambda(sampling, output_shape = (2,) )([z_mean, z_log_var])  #normal sampling layer
h_VAE = Input(shape=(2,))

dec_2_VAE = Dense(256,activation='relu')
dec_1_VAE = Dense(28*28, activation='relu')
```

```python
VAE_total = Model(x,dec_1_VAE(dec_2_VAE(NS_sampling)))
VAE_total.summary()
```

### 訓練完權重後 load_weights

```python
VAE_total.compile(loss = vae_loss, optimizer=Adam(), metrics=['acc'])
VAE_total.load_weights('./VAE_MNIST_model/Model_2nd_weights.h5')
#Model_2nd_history = VAE_total.fit(x_train,x_train, batch_size = 1024, epochs = 128)
```

```python
#Model_2nd_json = VAE_total.to_json()
#open('./VAE_MNIST_model/Model_2nd','w').write(Model_2nd_json)
#VAE_total.save_weights('./VAE_MNIST_model/Model_2nd_weights.h5')
```

### 把 encoder 跟 decoder 分開

```python
VAE_mean_encoder = Model(x, z_mean)
VAE_mean_encoder.summary()
```

```python
VAE_decoder = Model(h_VAE, dec_1_VAE(dec_2_VAE(h_VAE)))
VAE_decoder.summary()
```

### 可視化 encoded 後的二維變數，用了 VAE+relu 後，可以看到不同數字之間的界線變得更清晰，而且同樣數字的資料點變得比較群聚。

```python
sample = np.random.randint(len(x_test), size=len(x_test)//2)
encoder_vis = VAE_mean_encoder.predict(x_test[sample])
plt.scatter(encoder_vis[:,0], encoder_vis[:,1], c = y_test0[sample], cmap='tab10')
plt.colorbar()
plt.xlabel('$x_{mean}$')
plt.ylabel('y_{mean}')
plt.show()
```

### 可視化指定的二維平面上各點被 decoded 後的結果，分佈也與上圖類似

```python
n = 32
pixel_side_each = 28
pixel_side_total = pixel_side_each*n
fig = np.zeros((pixel_side_total,pixel_side_total))
grid_x = np.linspace(-4.25, 4.25, n)
grid_y = np.linspace(-4.25, 4.25, n)

for i in range(0,n):
    for j in range(0,n):
        input_data = np.array([[grid_x[i], grid_y[j]]])
        temp = VAE_decoder.predict(input_data)
        temp = temp[0].reshape(28,28)
        temp = 255*(temp-temp.min())/temp.max()
        #print(temp.max())
        fig[(n-j-1) * pixel_side_each: (n-j) * pixel_side_each, #from up to down
            i * pixel_side_each: (i + 1) * pixel_side_each] = temp # from left to right
```

```python
plt.figure(figsize=(20, 20))
plt.imshow(fig, cmap='Greys')
plt.axis('off')
plt.show()
```

### 用 interact_manual 讓使用者挑選 training data中的兩張圖，用這兩張圖 encoded 後得到的兩組參數， 在二維平面上決定一個矩形，並把四個角的參數 decoded 後的圖畫出來。接著再根據使用者指定的 t_x, t_y 找到矩形中對應的點，分別畫出 decoded 後的圖案並與直接使用四個角的圖做疊合得到的圖比較。最後畫出該點在二為平面中的位置。

```python
def DRAW_IN_BETWEEN_POINT(index_1, index_2, t_x, t_y):
    mean_1 = VAE_mean_encoder.predict(x_train[index_1:index_1+1])[0]
    mean_2 = VAE_mean_encoder.predict(x_train[index_2:index_2+1])[0]
    plt.figure(figsize=(15,10))
    FONTSIZE = 8

    if mean_1[0]>= mean_2[0]:
        x_max = mean_1[0]
        x_min = mean_2[0]
    else:
        x_max = mean_2[0]
        x_min = mean_1[0]
    if mean_1[1]>= mean_2[1]:
        y_max = mean_1[1]
        y_min = mean_2[1]
    else:
        y_max = mean_2[1]
        y_min = mean_1[1]
    if (mean_1[0]-mean_2[0])*(mean_1[1]-mean_2[1])>0:
        ax_1 = plt.subplot(3,6,6)
        ax_2 = plt.subplot(3,6,13)
    else:
        ax_1 = plt.subplot(3,6,18)
        ax_2 = plt.subplot(3,6,1)
    if mean_1[0]>=mean_2[0]:
        ax_1.imshow(x_train0[index_1], cmap='Greys')
        ax_1.axis('off')
        ax_1.set_title(f'Real Hand-written Number \nat ({mean_1[0]:.2f}, {mean_1[1]:.2f})', fontsize=FONTSIZE)
        ax_2.imshow(x_train0[index_2], cmap='Greys')
        ax_2.axis('off')
        ax_2.set_title(f'Real Hand-written Number \nat ({mean_2[0]:.2f}, {mean_2[1]:.2f})', fontsize=FONTSIZE)
    else:
        ax_2.imshow(x_train0[index_1], cmap='Greys')
        ax_2.axis('off')
        ax_2.set_title(f'Real Hand-written Number \nat ({mean_1[0]:.2f}, {mean_1[1]:.2f})', fontsize=FONTSIZE)
        ax_1.imshow(x_train0[index_2], cmap='Greys')
        ax_1.axis('off')
        ax_1.set_title(f'Real Hand-written Number \nat ({mean_2[0]:.2f}, {mean_2[1]:.2f})', fontsize=FONTSIZE)

    delta_x = x_max-x_min
    delta_y = y_max-y_min
    x_interact = x_min + delta_x*t_x
    y_interact = y_min + delta_y*t_y

    plot_index = [2,5,14,17,9]

    input_data = np.array([
                    [x_min, y_max],
                    [x_max, y_max],
                    [x_min, y_min],
                    [x_max, y_min],
                    [x_interact, y_interact]
                ])
    temp = VAE_decoder.predict(input_data)
    for i,element in enumerate(temp):
        element = (element-element.min())/element.max()
        element = element.reshape(28,28)
        ax = plt.subplot(3, 6, plot_index[i])
        ax.imshow(element, cmap='Greys')
        ax.axis('off')
        ax.set_title(f'VAE decoder Generated Number\n at ({input_data[i][0]:.2f}, {input_data[i][1]:.2f})', fontsize=FONTSIZE)


    overlap = (1-t_x)*t_y*temp[0].reshape(28,28)+t_x*t_y*temp[1].reshape(28,28)+(1-t_x)*(1-t_y)*temp[2].reshape(28,28)+t_x*(1-t_y)*temp[3].reshape(28,28)
    ax = plt.subplot(3,6,10)
    ax.imshow(overlap, cmap='Greys')
    ax.axis('off')
    ax.set_title(f'Superposition of Four Corners\n at ({input_data[-1][0]:.2f}, {input_data[-1][1]:.2f})', fontsize=FONTSIZE)

    figure = plt.figure(figsize=(5,5))
    plt.xlabel('$x_{mean}$')
    plt.ylabel('$y_{mean}$')
    #plt.xlim(-0.5*delta_x+x_min,0.5*delta_x+x_max)
    #plt.xticks(np.arange(-4,5,1))
    #plt.ylim(-0.5*delta_y+y_min,0.5*delat_y+y_max)
    #plt.yticks(np.arange(-4,5,1))
    plt.hlines(y=y_min, xmin=x_min, xmax=x_max, linewidth=2, color ='r', linestyles='dashed')
    plt.hlines(y=y_max, xmin=x_min, xmax=x_max, linewidth=2, color ='r', linestyles='dashed')
    plt.vlines(x=x_min, ymin=y_min, ymax=y_max, linewidth=2, color ='r', linestyles='dashed')
    plt.vlines(x=x_max, ymin=y_min, ymax=y_max, linewidth=2, color ='r', linestyles='dashed')
    plt.plot([mean_1[0],mean_2[0]], [mean_1[1],mean_2[1]],'r.', markersize=20)
    plt.text(mean_1[0]+0.03*delta_x, mean_1[1]+0.03*delta_y,'Index 1',color='r')
    plt.text(mean_2[0]+0.03*delta_x, mean_2[1]+0.03*delta_y,'Index 2',color='r')
    plt.plot(x_interact, y_interact,'b.', markersize=20)
    plt.text(x_interact+0.03*delta_x, y_interact+0.03*delta_y,'Position of in-between Point',color='b')

    plt.show()
```

```python
interact_manual(DRAW_IN_BETWEEN_POINT, 
             index_1 = IntSlider(30829,0,len(x_train)), 
             index_2 = IntSlider(44444,0,len(x_train)), 
             t_x=FloatSlider(value = 0.5, min = 0, max = 1.0, step = 0.02), 
             t_y=FloatSlider(value = 0.5, min = 0, max = 1.0, step = 0.02));
```
