```python
###############################################################################
def fill_check(number):
    i=0
    while i<9:
        j = 0
        while j<9:
            if number[i][j] == ' ':
                break
            else:
                j+=1
        if(j!=9):
            break
        else:
            i+=1
    if  j==9 and i==9 :
        return(True)
    else:
        return(False)
###############################################################################
def board_display():
    str1='* * * * * * * * * * * * * * * * * * *'
    str2='====================================='
    board=str1    
    copy = [['' for i in range(0,9)] for j in range(0,9)]
    for i in range(0,9):
        for j in range(0,9):
            if (i,j) in fixed_block:
                copy[i][j] = '\033[94m' + '\033[1m' + str(number[i][j]) + '\033[0m'
            else:
                copy[i][j] = str(number[i][j])
               
    for i in range (0,9):
        board += '\n*'
        for j in range (0,9):
            if (j%3==2 and j!=8):
                board += ' '+ copy[i][j] + ' ' + '|'
            else:
                board += ' '+ copy[i][j] + ' ' + '*'
        if (i%3==2 and i!=8):
            board += '\n' + str2
        else:    
            board += '\n' + str1
    print('\n'+board)
###############################################################################
def answer_check():    
    answer=[[5, 3, 4, 6, 7, 8, 9, 1, 2],
     [6, 7, 2, 1, 9, 5, 3, 4, 8],
     [1, 9, 8, 3, 4, 2, 5, 6, 7],
     [8, 5, 9, 7, 6, 1, 4, 2, 3],
     [4, 2, 6, 8, 5, 3, 7, 9, 1],
     [7, 1, 3, 9, 2, 4, 8, 5, 6],
     [9, 6, 1, 5, 3, 7, 2, 8, 4],
     [2, 8, 7, 4, 1, 9, 6, 3, 5],
     [3, 4, 5, 2, 8, 6, 1, 7, 9]]
    i=0;
    while i<9:
        if (number[i]!=answer[i]):
            break
        else:
            i+=1
    if i==9:
        return(True)
    else:
        return(False)
###############################################################################
def answer_show():
    answer=[[5, 3, 4, 6, 7, 8, 9, 1, 2],
     [6, 7, 2, 1, 9, 5, 3, 4, 8],
     [1, 9, 8, 3, 4, 2, 5, 6, 7],
     [8, 5, 9, 7, 6, 1, 4, 2, 3],
     [4, 2, 6, 8, 5, 3, 7, 9, 1],
     [7, 1, 3, 9, 2, 4, 8, 5, 6],
     [9, 6, 1, 5, 3, 7, 2, 8, 4],
     [2, 8, 7, 4, 1, 9, 6, 3, 5],
     [3, 4, 5, 2, 8, 6, 1, 7, 9]]
    str1='* * * * * * * * * * * * * * * * * * *'
    str2='====================================='
    
    board=str1
    copy = [['' for i in range(0,9)] for j in range(0,9)]
    for i in range(0,9):
        for j in range(0,9):
            if (i,j) in fixed_block:
                copy[i][j] = '\033[94m' + '\033[1m' + str(answer[i][j]) + '\033[0m'
            elif answer[i][j]!=number[i][j]:
                copy[i][j] = '\033[91m' + '\033[1m' + str(answer[i][j]) + '\033[0m'
            else:
                copy[i][j] = str(answer[i][j])
               
    for i in range (0,9):
        board += '\n*'
        for j in range (0,9):
            if (j%3==2 and j!=8):
                board += ' '+ copy[i][j] + ' ' + '|'
            else:
                board += ' '+ copy[i][j] + ' ' + '*'
        if (i%3==2 and i!=8):
            board += '\n' + str2
        else:    
            board += '\n' + str1
    print('\n'+board)
###############################################################################
def restart():
    for i in range(0,9):
        for j in range(0,9):
            if (i,j) not in fixed_block:
                number[i][j] = ' '
    for i in range(0,4):
                status[i]=False
    record[i] = number[i].copy()
###############################################################################
def command_identifier(command):
    # success,saved,quit,fill 1 2 3
    if status[2] and command!='restart':
        print('\n掰掰~可按 \033[91my\033[0m 看答案或 \033[91mrestart\033[0m 重新開始喔~(*´∀`)~♥')
        if command == 'y':
            clear_output(wait=True)
            print('答案是:')
            answer_show()
            print('可以輸入 \033[91mrestart\033[0m 再挑戰一次喔~') 
    elif status[0] and command!='restart':
        print('\n按 \033[91mrestart\033[0m 重新開始喔~(*´∀`)~♥')
    elif command == 'start':
        print('\n等待指令中...(ﾟ3ﾟ)～♪')
    elif command == 'restart':
        restart()
    elif command == '':
        print('您沒有輸入東西喔 (ㆆᴗㆆ)')
    elif len(command)==1:
        if command[0] == 's':
            for i in range(0,9):
                record[i] = number[i].copy()
            print('\n存檔唷~')
            status[1] = True
        elif command[0] == 'r':
            if (status[1]):
                for i in range(0,9):
                    number[i] = record[i].copy()
                print('\n讀檔唷~')
            else:
                print('\n您還沒存過檔喔~')
        elif command[0] == 'q':
            status[2] = True
            print('\n您不玩了嗎? Miss you ~QWQ~')
            print('想知道答案嗎?請按 \033[91my\033[0m 看答案,或其他鍵離開喔 (๑•́ ₃ •̀๑)\n')         
        elif command[0] == 'a':
            if status[3] == True:
                status[0] = answer_check()
                if status[0] == True:
                    print('\n\033[94m正解喔!!恭喜您脫離苦海 ｡:.ﾟヽ(*´∀`)ﾉﾟ.:｡ \033[0m')
                    print('\n可以輸入 \033[91mrestart\033[0m 再挑戰一次喔(O.S.:何必呢~(´,_ゝ`))')
                else:
                    print('\n答案不對喔~ლ(・´W`・ლ) 再檢查一下吧!')
            else:
                print('\n填滿再叫我_(:3 」∠ )_')
        else:
            print('\n您有閱讀上的問題嗎? OwOa 請再輸一次指令')
    else:
        if len(command) == 5 and command[0] == 'w' and command[3] == '=':
            if (command[1] in allow and command[2] in allow)==True:
                row = int(command[1])-1
                column = int(command[2])-1
                if (row,column) in fixed_block:
                        print('\n亂動題目是不可以的喔~歐逆/內醬~(*´艸`*) 請再輸一次指令')
                else:
                    value = command[4]
                    if value in allow:
                        number[row][column] = int(value)
                        print(f'\n把第{row+1}列第{column+1}格填成{value}囉~')
                    else:
                        print('您知道數獨嗎? OwOa 請重填想輸入的數字~')
            else:
                print('您有認真嗎? OwOa 輸入的列數或格數有問題喔~')
        elif len(command) == 3 and command[0] == 'e':
            if (command[1] in allow and command[2] in allow)==True:
                row = int(command[1])-1
                column = int(command[2])-1
                if (row,column) in fixed_block:
                    print('\n亂動題目是不可以的喔~歐逆/內醬~(*´艸`*) 請再輸一次指令')
                elif number[row][column] == ' ':
                    print('\n這格沒東西唷~免擦免擦_(°ω°｣ ∠) 請再輸一次指令')
                else:
                    number[row][column] = ' '
                    print(f'\n把第{row+1}列第{column+1}格擦掉囉~')
        else:
            print("\n您有閱讀上的問題嗎? OwOa 請再輸一次指令")
    return(status)
###############################################################################
def sudoku(command):
    clear_output(wait=True)  
    status=command_identifier(command)
    
    if status[0]==False:          
        status[3] = fill_check(number)
        
        if not status[2]:
            print('''\n\033[92m指令參照表(列數及格數都從1起跳喔~):
        w12=3 => 把第1列第2格填入/改成3 3前面要有空格,填的數字只能是1到9喔~
        e45 => 把第4列第5格擦掉
        s => 存檔
        r => 讀檔
        a => 對答案 只有填滿所有空格時才能用喔
        q => 退出並選擇是否顯示解答
        restart => 重新開始 存檔會被洗掉喔
列數跟行數只能是1到9,剩下不行喔~\033[0m
快輸入指令吧 OwO/\n''')

        if status[3] == True:
            print('\n\033[94m***恭喜您填完囉ε٩(๑> ₃ <)۶з~可以按 a 對答案,或是用 e 或 w 指令繼續更改唷~***\033[0m\n')                     

        board_display()
    else:     
        if command == 'restart':
            restart()
            board_display()
###############################################################################
```
